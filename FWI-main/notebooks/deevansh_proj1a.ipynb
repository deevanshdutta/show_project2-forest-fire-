```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv",header=1)
```

```python
df.info()
```

```python
##data cleaning
##missing values
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
##by observing dataset sidi bel region start is 122 and before it is belgia
```

```python
##so splitting dataset into regions let region=0 nd 1
```

```python
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1##sidi bel region
```

```python
df.info()
```

```python
df[["Region"]]=df[["Region"]].astype(int)
```

```python
df.info()
```

```python
df.tail()
```

```python
df.isnull().sum()
```

```python
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.columns
```

```python
##fix spaces in column name
```

```python
df.columns=df.columns.str.strip()
```

```python
df.columns
```

```python
df.head()
```

```python
##changing other column to float 
```

```python
objects=[feature for feature in df.columns if df[feature].dtypes=='O']
```

```python
objects
```

```python
##except classes , change other datatypes also
```

```python
for i in objects:
    if i!="Classes":
        df[i]=df[i].astype(float)
```
